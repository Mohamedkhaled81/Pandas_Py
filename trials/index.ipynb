**The core difference:**
- `.loc` selects by **label** (the actual index value)
- `.iloc` selects by **integer position** (0, 1, 2... regardless of what the labels are)

When your DataFrame has the default index (0, 1, 2, 3...), they often *look* the same, which is why the distinction doesn't show up until you set a custom index.

**Example — this is where it gets interesting:**

```python
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Carol'],
    'score': [90, 85, 95]
})

df = df.set_index('name')
print(df)
#        score
# name
# Alice     90
# Bob       85
# Carol     95
```

Now:

```python
df.loc['Bob']      # works — 'Bob' is a label in the index
df.iloc[1]         # also works — position 1 (second row)

df.loc[1]          # ❌ KeyError — there's no label "1" anymore
```

Once you `set_index('name')`, the integer position `0, 1, 2` no longer exists as a label — so `.loc[1]` fails, but `.iloc[1]` still works because it just counts rows.

**The classic gotcha — when your index is *also* integers but not sequential:**

```python
df2 = pd.DataFrame({'val': [10, 20, 30]}, index=[5, 3, 8])

df2.loc[3]    # returns 20 — looks up the label "3"
df2.iloc[3]   # ❌ IndexError — there's no position 3, only 0,1,2 exist
```

This is the case that really separates the two: `df2.loc[1]` looks for a *label* equal to 1, while `df2.iloc[1]` always means "the second row," no matter what labels are sitting in the index.

**Quick rule of thumb:**
| | `.loc` | `.iloc` |
|---|---|---|
| Uses | Labels (names, dates, custom index values) | Integer positions (always 0-based) |
| Slicing `[a:b]` | **Inclusive** of `b` | **Exclusive** of `b` (like normal Python) |
| Fails when | Label doesn't exist | Position is out of range |

That slicing difference is worth remembering too: `df.loc['Alice':'Bob']` includes Bob, but `df.iloc[0:1]` stops *before* index 1.
